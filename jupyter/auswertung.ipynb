```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.optimize import curve_fit
import statsmodels.api as sm
import statistics as stat
import numpy as np
```

```python
stats = pd.read_csv('../resources/stats/stats.csv')
sums = stats.groupby('game').sum().reset_index()
```

```python
# Belohnungen pro Spiel aller Agenten mit LOWESS-Regression

sums_teams = (sums['reward_1'] + sums['reward_2'] + sums['reward_3'] + sums['reward_4']).reset_index()
sums_teams.plot(kind='scatter', x='index', y=0, xlabel='Spiel')
lowess = sm.nonparametric.lowess(sums_teams[0], sums_teams['index'], frac=0.2)
plt.plot(lowess[:, 0], lowess[:, 1], color='red')
plt.ylabel("B        ", rotation=0)
plt.savefig('img/img_lowess.png')
plt.show()
print(f'Arithmetisches Mittel der Summe aller Belohnungen: {sums_teams.mean()[0]}')
print(f'Standardabweichung: {sums_teams.std()[0]}')
```

```python
lowess = sm.nonparametric.lowess(sums['pos_1'], sums['game'], frac=0.1)
lowess1 = sm.nonparametric.lowess(sums['pos_2'], sums['game'], frac=0.1)

dlowess = np.gradient(lowess[:, 1], lowess[:, 0])
dlowess1 = np.gradient(lowess1[:, 1], lowess1[:, 0])

correlation = np.corrcoef(dlowess, dlowess1)[0, 1]

print(correlation)
```
